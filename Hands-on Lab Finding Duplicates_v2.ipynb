### Task 2: Analyze Characteristics of Duplicates
1. Identify which columns have the same values in duplicate rows.
2. Analyze the distribution of duplicates across different columns such as Country, Employment, and DevType.


```python
from IPython.display import HTML, display

def find_duplicate_columns():
    # Get all duplicate rows
    duplicate_rows = df[df.duplicated(keep=False)]
    display(HTML(f"Total Duplicate Rows <span style='color: blue'><b>{duplicate_counts}</b></span>"))
    
    # Compare values in each column
    for column in df.columns:
        # Check if values in this column are identical for duplicate rows
        if duplicate_rows[column].nunique() == 1:
            display(HTML(f"Column <span style='color: blue'><b>{column}</b></span> has identical values: <span style='color: green'><b>{duplicate_rows[column].iloc[0]}</b></span>"))

# Run the analysis
find_duplicate_columns()
```

```python
def display_results_table(results_dict):
    # Create HTML table with styling
    html_table = """
    <table style="width:auto; border-collapse: collapse;">
        <tr style="background-color: #4CAF50; color: white;">
            <th style="text-align: left;">Category</th>
            <th style="text-align: center;">Value</th>
            <th style="text-align: left;">Count</th>
            <th style="text-align: left;">Unique Values</th>
        </tr>
    """
    # Add rows with alternating colors
    for column in results_dict:
        for idx, (value, count) in enumerate(results_dict[column]['counts'].items()):
            bg_color = 'light gray' if idx % 2 == 0 else 'gray'
            html_table += f"""
            <tr style="background-color: {bg_color};">
                <td style="text-align: left;">{column}</td>
                <td style="text-align: left;"><span style="color:Yellow;">{value}</span></td>
                <td ><span style="color:Yellow;">{count}</span></td>
                <td >{results_dict[column]['unique_values']}</td>
            </tr>
            """
    html_table += "</table>"
    display(HTML(html_table))
```

```python
def analyze_duplicate_distribution(df, columns_to_check):
    # Dictionary to store results
    results_dict = {}    
    for column in columns_to_check:
        # Get duplicate counts and store in dictionary
        duplicate_counts = df[df.duplicated(subset=[column], keep=False)][column].value_counts()
        results_dict[column] = {
            'counts': duplicate_counts.head(-1),  # Store top 10 duplicates
            'unique_values': df[column].nunique()
        }
    return results_dict

# Run analysis and get results
columns_to_check=['Country', 'Employment', 'DevType']
results = analyze_duplicate_distribution(df,columns_to_check)
print("Results Dictionary:")
display_results_table(results)
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
for i, column in enumerate(columns_to_check[2:], 1):
    plt.figure(figsize=(12, 8))
    plt.subplot(len(columns_to_check)-2, 1, i)
    results[column]['counts'].head(10).plot(kind='pie', autopct='%1.1f%%')
    plt.title(f'Top 10 Duplicates in {column}\nTotal Unique Values: {results[column]["unique_values"]}')
    plt.xticks(rotation=90)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for i, column in enumerate(columns_to_check[:2], 1):
    plt.figure(figsize=(12, 8))
    plt.subplot(len(columns_to_check)-1, 1, i)
    results[column]['counts'].head(20).plot(kind='bar')
    plt.title(f'Top 20 Duplicates in {column}\nTotal Unique Values: {results[column]["unique_values"]}')
    plt.xticks(rotation=90)
    plt.ylabel('Frequency')
    plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
def analyze_and_clean_duplicates(df, key_columns):
    # 1. Create HTML list items for each column and its rationale
    column_list_items = ''.join([f'<li><b>{col}:</b> {reason}</li>' for col, reason in key_columns.items()])
    
    # 2. Initial Analysis and Documentation
    display(HTML(f"""
    <h2 style='color: #1976D2'>Duplicate Analysis Process</h2>
    <h3 style='color: #2196F3'>1. Critical Columns Selected:</h3>
    <ul>
        {column_list_items}
    </ul>
    """))
    
    # 3. Document duplicate statistics
    initial_stats = {
        'Total Records': len(df),
        'Duplicate Rows': df.duplicated(subset=key_columns.keys()).sum(),
        'Unique Combinations': df.groupby(list(key_columns.keys())).size().shape[0]
    }
    
    display(HTML(f"""
    <h3 style='color: #2196F3'>2. Initial Statistics:</h3>
    <ul>
        <li>Total Records: <b>{initial_stats['Total Records']}</b></li>
        <li>Duplicate Rows: <b>{initial_stats['Duplicate Rows']}</b></li>
        <li>Unique Combinations: <b>{initial_stats['Unique Combinations']}</b></li>
    </ul>
    """))
    
    # 4. Remove duplicates and show results
    df_clean = df.drop_duplicates(subset=list(key_columns.keys()))
    
    display(HTML(f"""
    <h3 style='color: #2196F3'>3. Results After Cleaning:</h3>
    <ul>
        <li>Records Removed: <b>{len(df) - len(df_clean)}</b></li>
        <li>Final Record Count: <b>{len(df_clean)}</b></li>
        <li>Duplicate Removal Rate: <b>{((len(df) - len(df_clean))/len(df)*100):.2f}%</b></li>
    </ul>
    """))
    return df_clean

print("___________________________________________________Use Case I (Country,Employment and DevType) ____________________________________________________")
key_columns_1 = {'Country':'Geographic distribution is crucial for demographic analysis', 'Employment':'Primary identifier for employment status', 'DevType': 'Key professional categorization'}
clean_df = analyze_and_clean_duplicates(df,key_columns_1)

print("___________________________________________________Use Case II (Country,Employment,DevType and Industry) ____________________________________________________")

key_columns_2 = {'Country':'Geographic distribution is crucial for demographic analysis', 'Employment':'Primary identifier for employment status', 'DevType': 'Key professional categorization','Industry':'Domain Expert'}
clean_df = analyze_and_clean_duplicates(df,key_columns_2)
```
